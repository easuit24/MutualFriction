```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

from gpekohnmode import GPETimeEv as gpe
```

```python
g = gpe(dim = 2)  # try to get the animation to work here using matplotlib widget if not, make sure the plot is changing... 
```

Notes to come back to: KE and PE seemed to be oscillating, but the time window was too short to really see, so I extended the time window out to see if KE and PE are oscillating as expected for Kohn's modes. 

```python
print(g.g) 
```

```python
plt.figure() 
plt.imshow(np.abs(g.psi) , origin = 'lower') 
plt.show() 
```

```python
plt.imshow(np.abs(g.dynpsi) ) 
plt.show() 
```

```python
#%matplotlib widget
plt.figure()
plt.plot(np.array(g.Ep_arr)/g.Natoms, label = 'Potential Energy') 
plt.plot(np.array(g.Ek_arr)/g.Natoms, label = 'Kinetic Energy') 
plt.plot(np.array(g.Ei_arr)/g.Natoms, label = 'Interaction') 
plt.plot(np.array(g.Ep_arr)/g.Natoms + np.array(g.Ek_arr)/g.Natoms + np.array(g.Ei_arr)/g.Natoms, label = 'Total')
plt.plot(np.array(g.Ep_arr)/g.Natoms - np.array(g.Ek_arr)/g.Natoms , label = 'Virial')
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend() 
plt.title("Kohn Modes: Energy Profile for well displacement = 7, small interactions")
#plt.savefig('kohnenergy1.png')
```

```python
plt.figure()

plt.plot(np.array(g.Ei_arr)/g.Natoms, label = 'Interaction') 
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend() 
plt.title("Kohn Modes: Interaction Energy")
#plt.savefig('kohnenergy1.png')
```

```python
%matplotlib widget
fig, ax = plt.subplots() 
data = plt.imshow(np.abs(g.snapshots[0]), extent = [-g.winL/2, g.winL/2, -g.winL/2, g.winL/2],cmap = plt.cm.hot, origin = 'lower') 
plt.xlabel("x", fontsize = 16)
plt.ylabel('y', fontsize = 16)
plt.contour(g.xi[0], g.xi[1], g.Vs)
plt.title(f'Animation for w={g.w}, L={g.L}')

def animate(i): 
    data.set_data(np.abs(g.snapshots[i]))
    return [data]
anim = animation.FuncAnimation(fig, animate, frames = len(g.snapshots))
anim.save(f'kohn_2.mp4')
plt.show()
```
