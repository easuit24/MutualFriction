In this notebook, we aim to replicate the experimental results of J. Dalibard (PRX 9, 021035) through theory. First we attempt to replicate the initial result, where it is observed that the Bose-Einstein Condensate (BEC) transitions from a uniform, square density to a spherically symmetric one upon switching the trap from an optical box to a harmonic magnetic trap. This behavior corresponds to a switch from a box potential to a harmonic one. Experimentally in this transition, the atoms experience two consecutive microwave transitions to make them field-sensitive (F = 1, $m_f = -1$). It is also observed that the potential energy is oscillatory with frequency $2\omega$ and can be fitted with a cosine function.
