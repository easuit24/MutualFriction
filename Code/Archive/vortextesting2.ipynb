```python
import numpy as np
import matplotlib.pyplot as plt

from gpevortex import GPETimeEv as gpev
```

```python
g = gpev(L = 50, dtcoef = 0.05, dim = 2, numImagSteps=2000, numRealSteps = 108000, dist = 3, antiV = True)
```

```python
%matplotlib widget
g.animatepsi2d_show() 
```

```python
psi_inter = g.dynpsi
```

```python
%matplotlib widget
plt.figure() 
plt.imshow(np.abs(psi_inter), cmap = plt.cm.hot, extent = [-g.winL/2, g.winL/2,-g.winL/2,g.winL/2])
plt.xlabel("X")
plt.ylabel("Y") 
plt.title("Intermediate Wavefunction")
plt.show() 
```

```python
g = gpev(L = 50, dtcoef = 0.05, dim = 2, numRealSteps = 108000, dist = 3, antiV = True, imp = True, impPsi = psi_inter)
```

```python
g.animatepsi2d_show() 
```

```python
g.animatepsi2d_save()
```
