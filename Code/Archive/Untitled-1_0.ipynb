```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from classicalfield_orig import FiniteTempGPE as gpe
```

```python
g = gpe(npoints = 2**6, numImagSteps = 2000, numRealSteps = 1000, dtcoef = 0.0005, boxthickness = 0.4, Nsamples = 1, runAnim = True, animFileName='test', Tfact = 1, dst = False, vortex = False)
```

```python
potential_energy = g.gpeobj.Vbox
```

```python
plt.figure(figsize = (6,6))
ax = plt.gca() 
plt.imshow(potential_energy, extent = [-g.winL/2, g.winL/2, -g.winL/2, g.winL/2], cmap = plt.cm.hot)

plt.xlabel('X')
plt.ylabel('Y')
divider = make_axes_locatable(ax) 

cax = divider.append_axes("right", size = "5%", pad = 0.1)
plt.colorbar(label = 'Potential Energy', cax = cax)
```
